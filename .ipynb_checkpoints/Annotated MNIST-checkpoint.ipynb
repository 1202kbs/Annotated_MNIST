```python
import cv2

def rotate_and_scale(img, angle, scale):
    M = cv2.getRotationMatrix2D((14, 14), angle, scale)
    return cv2.warpAffine(img, M, (28, 28))
```

```python
col = []
for angle in range(-90, 90, 5):
    temp1 = rotate_and_scale(images[h_idx[18]], angle, 1)
    temp2 = temp1[:, 12:16]
    col.append(np.sum(temp2 != 0))

idx2ang = {idx: angle for (idx, angle) in enumerate(range(-90, 90, 5))}
idx = np.argmax(col, axis=0)
print('Angle:', idx2ang[idx], ' Value:', col[idx])
```

```python
np.mean(col)
```
